```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import metrics
%matplotlib inline
```

```python
df = pd.read_csv("C:\c++\My codes\csv files\Mall_Customers.csv")
df
```

```python
x = df.iloc[:, 3:]
```

```python
x
```

```python
wcss = []
for i in range(1, 15):
    kmean = KMeans(n_clusters=i, init='k-means++', random_state=30)
    kmean.fit(x)
    wcss.append(kmean.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 15), wcss)
```

```python
x
```

```python
kmean1 = KMeans(n_clusters=5, init='k-means++', random_state=30)
kmean1.fit_predict(x)
```

```python
x['cluster number'] = kmean1.fit_predict(x)
```
